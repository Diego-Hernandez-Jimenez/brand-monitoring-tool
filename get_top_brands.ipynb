```python
import requests
from bs4 import BeautifulSoup
```

```python
target = 'https://www.merca20.com/the-500-most-valuable-brands-according-to-the-brand-finance-2025-ranking/'
response = requests.get(target)
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
list_table = soup.find('table', attrs={'cellspacing': '0', 'cellpadding':'0'})
```

```python
table_contents = list_table.find_all('td')
top_brands = [table_contents[i].text for i in range(6, len(table_contents), 4)]

top_brands[top_brands.index('Facebook')] = 'Facebook Meta'
top_brands[top_brands.index('TikTok/Douyin')] = 'TikTok'
top_brands
```
